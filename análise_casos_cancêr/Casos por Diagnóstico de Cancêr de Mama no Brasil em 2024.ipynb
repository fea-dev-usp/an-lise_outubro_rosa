```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando o arquivo Excel e exibindo o DataFrame 
# O parâmetro index_col=0 define a primeira coluna como índice do DataFrame
dados = pd.read_excel(r"Casos por Diagnóstico Detalhado segundo UF da residência.xlsx", index_col=0)
```

```python
dados
```

## Total de Casos Segundo UF da Residência

```python
# Removendo a última linha
dados = dados.drop(['Total'])
```

```python
#Plotando o gráfico
dados.plot(kind='barh', figsize=(10, 6), color= "hotpink")

# Definir título e rótulos
plt.title('Casos segundo UF de residencia em 2024')
plt.xlabel('Casos')
plt.ylabel('')
```

O gráfico mostra que São Paulo tem uma incidência consideravelmente maior que os demais, destacando-se como o estado com maior número de casos, passando de 6.000 e outros estados com números expressivos de casos incluem Bahia, Paraná, Pernambuco, Rio de Janeiro e Rio Grande do Sul. Em contrapartida, estados como Acre, Amapá, Roraima e Tocantins apresentam poucos casos, possivelmente abaixo de 100.

##  Casos Segundo Região do Brasil  em 2024##

### Região Norte ###

```python
# DataFrame contendo apenas os dados dos estados do Norte
norte = ['Acre', 'Amapá', 'Amazonas', 'Pará', 'Rondônia', 'Roraima', 'Tocantins']

dados_norte = dados.loc[norte]
dados_norte
```

```python
#Plotando o gráfico
dados_norte.plot(kind='bar', figsize=(10,6), color= "hotpink")

# Definir título e rótulos
plt.title('Casos de Cancêr de Mama na Região Norte')
plt.xlabel('')
plt.xticks(rotation=360, fontsize=9)
plt.ylabel('Casos')
```

Podemos concluir que o estado do Pará tem o maior número de casos 350 aproximadamente, seguido pelo Amazonas e Rondônia, com aproximadamente 200 casos cada. Esses estados, por serem os mais populosos da região, desenvolveram uma infraestrutura de saúde que facilita o diagnóstico. Por outro lado, estados como Acre, Amapá, Roraima e Tocantins têm menos de 50 casos, refletindo não só a menor população, mas também as dificuldades de acesso ao diagnóstico e ao tratamento, especialmente em áreas mais afastadas. Esse cenário reforça a importância de expandir a infraestrutura de saúde e as campanhas de conscientização para garantir que todas as pessoas tenham acesso ao atendimento necessário.

Entre os estados do Nordeste, a Bahia vem em primeiro lugar com quase 1.500 casos, enquanto para Pernambuco, Rio Grande do Norte e Ceará os números também são bastante significativos. Esses estados apresentam mais casos, pois têm população maior e melhor sistema de saúde. Em outros estados relativamente menores, como Sergipe e Alagoas, os números são bem menores.
