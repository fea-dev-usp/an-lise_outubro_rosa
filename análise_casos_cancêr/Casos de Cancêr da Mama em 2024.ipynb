* Sistema de Informação Ambulatorial (SIA), através do Boletim de Produção Ambulatorial Individualizado (BPA-I) e da Autorização de Procedimento de Alta Complexidade; Sistema de Informação Hospitalar (SIH); Sistema de Informações de Câncer (SISCAN)

```python
#Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando o arquivo Excel e exibindo o DataFrame 
# O parâmetro index_col=0 define a primeira coluna como índice do DataFrame
dados = pd.read_excel(r"Casos por Região em 2024.xlsx", index_col=0)
dados
```

```python
# Plotando o gráfico
dados.plot(kind='bar', figsize=(10,6), color = ["darkviolet","deeppink", "hotpink", "palevioletred", "fuchsia"])

# Definir título e rótulos
plt.title('Casos de Neoplasia Maligna da Mama em 2024 por Região')
plt.xticks(rotation=360, fontsize=9)
plt.ylabel('Pacientes')
plt.xlabel('')
```

A partir do gráfico acima podemos observar que ao todo foram registrados no Brasil 26.970 casos da doença. A região Sudeste destaca-se como a que teve mais ocorrência de neoplasia maligna (cancêr) da mama, com 11801 casos registrados, o que pode ser explicado por diversos fatores, como maior densidade populacional, maior acesso aos serviços de saúde e maior conscientização sobre a doença.

A disparidade entre os casos registrados nas regiões do Brasil serve como um alerta para a necessidade de ações urgentes por parte das autoridades de saúde para garantir o acesso equitativo aos serviços de prevenção, diagnóstico e tratamento do câncer de mama em todo o país.
