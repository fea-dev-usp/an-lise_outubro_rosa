```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando o arquivo Excel e exibindo o DataFrame 
# O parâmetro index_col=0 define a primeira coluna como índice do DataFrame
dados = pd.read_excel(r"Casos por Diagnóstico Detalhado segundo UF da residência.xlsx", index_col=0)
```

```python
dados
```

## Total de Casos Segundo UF da Residência

```python
# Removendo a última linha
dados = dados.drop(['Total'])
```

```python
#Plotando o gráfico
dados.plot(kind='barh', figsize=(10, 6), color= "hotpink")

# Definir título e rótulos
plt.title('Casos segundo UF de residencia em 2024')
plt.xlabel('Casos')
plt.ylabel('')
```

O gráfico mostra uma concentração significativa de diagnósticos no Sudeste, especialmente em São Paulo e Minas Gerais. Esse padrão provavelmente se deve à alta densidade populacional e à melhor infraestrutura de saúde, que facilita o acesso a exames de rotina. Além do Sudeste, os estados do Sul, como o Rio Grande do Sul, e do Nordeste, como a Bahia, também registram muitos diagnósticos, indicando uma boa capacidade de identificação da doença.

Por outro lado, estados do Norte e Centro-Oeste, como Acre, Amapá e Roraima, apresentam números baixos de casos, o que pode apontar para dificuldades de acesso a exames preventivos. Esses dados ressaltam a importância de ampliar o acesso a diagnósticos nessas regiões, garantindo que mais pessoas possam ser diagnosticadas e tratadas a tempo.








##  Casos Segundo Região do Brasil  em 2024 

### Região Norte ###

```python
# DataFrame contendo apenas os dados dos estados do Norte
norte = ['Acre', 'Amapá', 'Amazonas', 'Pará', 'Rondônia', 'Roraima', 'Tocantins']

dados_norte = dados.loc[norte]
dados_norte
```

```python
#Plotando o gráfico
dados_norte.plot(kind='bar', figsize=(10,6), color= "hotpink")

# Definir título e rótulos
plt.title('Casos de Cancêr de Mama na Região Norte')
plt.xlabel('')
plt.xticks(rotation=360, fontsize=9)
plt.ylabel('Casos')
```

Podemos concluir que o estado do Pará tem o maior número de casos 350 aproximadamente, seguido pelo Amazonas e Rondônia, com aproximadamente 200 casos cada. Esses estados, por serem os mais populosos da região, desenvolveram uma infraestrutura de saúde que facilita o diagnóstico. Os estados como Acre, Amapá, Roraima e Tocantins têm menos de 50 casos, refletindo não só a menor população, mas também as dificuldades de acesso ao diagnóstico e ao tratamento, especialmente em áreas mais afastadas. 



### Região Nordeste

```python
# DataFrame contendo apenas os dados dos estados do Nordeste
nordeste = ['Alagoas', 'Bahia', 'Ceará', 'Maranhão', 'Pará', 'Paraíba',
            'Pernambuco', 'Piauí', 'Rio Grande do Norte', 'Sergipe']

dados_nordeste = dados.loc[nordeste]
dados_nordeste
```

```python
#Plotando o gráfico
dados_nordeste.plot(kind='bar', figsize=(15,6), color= "hotpink")

# Definir título e rótulos
plt.title('Casos de Câncer de Mama na Região Nordeste')
plt.xlabel('')
plt.ylabel('Casos')
```

Entre os estados do Nordeste, a Bahia vem em primeiro lugar com quase 1.500 casos, enquanto para Pernambuco, Rio Grande do Norte e Ceará os números também são bastante significativos. 

### Região Centro-Oeste

```python
# DataFrame contendo apenas os dados dos estados do Centro-Oeste
centro_oeste = ['Goiás', 'Mato Grosso', 'Mato Grosso do Sul', 'Distrito Federal']

dados_centrooeste = dados.loc[centro_oeste]
dados_centrooeste
```

```python
#Plotando o gráfico
dados_centrooeste.plot(kind='bar', figsize=(10,6), color= "hotpink")

# Definir título e rótulos
plt.title('Casos de Cancêr de Mama na Região Centro-Oeste')
plt.xlabel('')
plt.xticks(rotation=360, fontsize=9)
plt.ylabel('Casos')
```

Podemos observar uma distribuição desigual dos casos de câncer de mama na região Centro-Oeste, com Goiás registrando significativamente mais casos, indicando possíveis diferenças no acesso a exames de detecção precoce, políticas de saúde pública, características demográficas, ou mesmo variáveis ambientais e socioeconômicas específicas de cada estado.

### Região Sudeste

```python
# DataFrame contendo apenas os dados dos estados do Sudeste
sudeste = ['Espírito Santo', 'Minas Gerais', 'São Paulo', 'Rio de Janeiro']

dados_sudeste = dados.loc[sudeste]
dados_sudeste
```

```python
#Plotando o gráfico
dados_sudeste.plot(kind='bar', figsize=(10,6), color= "hotpink")

# Definir título e rótulos
plt.title('Casos de Câncer de Mama na Região Sudeste')
plt.xlabel('')
plt.xticks(rotation=360, fontsize=9)
plt.ylabel('Casos')
```

Como mencionado anteriormente, São Paulo lidera em número de casos, ultrapassando 6.000. 

### Região Sul

```python
# DataFrame contendo apenas os dados dos estados do Sul
sul = ['Paraná', 'Santa Catarina', 'Rio Grande do Sul']

dados_sul = dados.loc[sul]
dados_sul
```

```python
#Plotando o gráfico
dados_sul.plot(kind='bar', figsize=(10,6), color= "hotpink")

# Definir título e rótulos
plt.title('Casos de Câncer de Mama na Região Sul')
plt.xlabel('')
plt.xticks(rotation=360, fontsize=9)
plt.ylabel('Casos')
```

Observa-se uma difereça de cerca de 700 casos entre os estados da região Sul. O Paraná e o Rio Grande do Sul lideram o número de casos, ambos com mais de 2.000 registros, o que aponta uma alta incidência de câncer de mama nessas regiões. Santa Catarina, embora também apresente um número expressivo, contabiliza menos de 1.500 casos.
