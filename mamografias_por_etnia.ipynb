```python
import pandas as pd

df = pd.read_csv("C:\\Users\\luane\\Downloads\\mamografia_residbr17294510762.csv", sep= ';')

print(df.head())

import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# configurar texto
index_to_drop = df[df['Raça/Cor'].str.lower() == 'total'].index
df = df.drop(index_to_drop)

index_to_drop2 = df[df['Raça/Cor'].str.lower() == 'indígena'].index
df = df.drop(index_to_drop2)

index_to_drop3 = df[df['Raça/Cor'].str.lower() == 'sem informação'].index
df = df.drop(index_to_drop3)

# montar gráfico
plt.figure(figsize=(10, 6))
plt.bar(df['Raça/Cor'], df['Exames'], color='pink')
plt.title('Número de Mamografias Realizadas por Etnia', fontsize=14)
plt.xlabel('Etnia', fontsize=12)
plt.ylabel('Número de Exames', fontsize=12)

# formatando eixo y para valores em milhões
def millions_formatter(x, pos):
    return f'{x/1e6:.1f}M'  
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter))


plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('grafico_mamografias.png')
plt.show()  
```
