# Total de Casos por Laudo de Mamografia em 2024

### Fonte

* Sistema de Informações de Câncer (SISCAN)

### Categorias dos Laudes de Mamografia

A divisão dos laudos em diferentes categorias é fundamental para a avaliação dos resultados dos exames de mamografia, na qual cada uma possui um significado clínico específico, orientando os profissionais de saúde sobre os próximos passos a serem tomados.

* Categoria 0: Incompleta ou não conclusiva;

* Categoria 1: Sem achados;

* Categoria 2:  Achados benignos (não são interpretados como suspeita de câncer);

* Categoria 3: Achados provavelmente benignos (exame deve ser repetido seis meses depois);

* Categoria 4: Achados suspeitos de malignidadeos (realização de uma biopsia do tecido suspeito, que indicará se existe ou não o risco de desenvolver o câncer);

* Categoria 5: Achados altamente suspeitos de malignidade;

* Categoria 6: Achados já com diagnóstico de câncer (utilizada apenas na mamografia diagnóstica, quando a mulher já tem
diagnóstico prévio de câncer de mama).

```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando o arquivo Excel e exibindo o DataFrame 
# O parâmetro index_col=0 define a primeira coluna como índice do DataFrame
dados = pd.read_excel(r"Pacientes por Laudo.xlsx", index_col=0)
```

```python
#Pegando dados apenas da última linha
dados_total = dados.loc['Total']

dados_total
```

```python
#Plotando o gráfico
dados_total.plot(kind='bar', figsize=(10, 6), color= ["violet", "fuchsia", "deeppink", "darkviolet", "hotpink", "mediumslateblue", "indigo"])

# Definir título e rótulos
plt.title('Total de Casos por Laudo de Mamografia em 2024')
plt.xlabel('')
plt.xticks(rotation=45, fontsize=9)
plt.ylabel('Pacientes')
```

O gráfico acima sugere que a maior parte dos laudos de mamografia em 2024 se concentrou nas categorias 2 e 1, sendo a categoria 2 predominante, ou seja, a maior parte dos laudos registrados indicam que não há suspeita de cancêr nos exames das pacientes. As categorias mais baixas (3 a 6) apresentam uma quantidade muito menor de laudos, possivelmente indicando que esses tipos de diagnósticos são menos frequentes ou que os pacientes com esses laudos representam uma minoria.
