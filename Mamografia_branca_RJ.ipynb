```python
#importar o mat 
import matplotlib.pyplot as plt

#colocar o conjunto de dados que vai plotar os dados 
labels = [2019, 2020, 2021, 2022, 2023]
vals = [30.225, 20.706, 39.563, 61.367, 97.637]

#construção dos gráficos
fig, ax = plt.subplots(figsize= (8, 8))

# tipo do gráfico
ax.bar(labels, vals)

#nome do titulo 1 em cima no meio
ax.set_title('Número de mamografias realizadas pela população branca no RJ', fontsize=13)
#nome do titulo 2 na vertical lado esquerdo
ax.set_ylabel( "Pacientes (valores entre 0 a 97.637 mil)" , fontsize=13 )
# titulo embaixo no meio
ax.set_xlabel( 'Ano' , fontsize=13 )


plt.show()
```
