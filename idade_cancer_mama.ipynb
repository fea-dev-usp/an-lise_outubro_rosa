```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('gbsg.csv')

# casos positivos (onde 'status' == 1)
df_positive = df[df['status'] == 1]

# encontrar a idade mais comum entre os casos positivos
most_common_age = df_positive['age'].mode()[0]
print(f"A idade mais comum de diagnóstico positivo é: {most_common_age} anos")

#plotar
plt.figure(figsize=(10, 6))
sns.histplot(df_positive['age'], kde=True, color='pink')
plt.title('Distribuição das Idades dos Diagnósticos Positivos de Câncer de Mama', fontsize=16)
plt.xlabel('Idade', fontsize=14)
plt.ylabel('Frequência', fontsize=14)

# idade mais comum no gráfico
plt.axvline(most_common_age, color='red', linestyle='--', label=f'Idade Mais Comum: {most_common_age}')
plt.legend()
plt.show()
```
